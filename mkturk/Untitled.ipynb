```python
import json
import pandas as pd
import numpy as np
```

```python
file_name = './active_hits-v1-argumentative-res-2379498895443566127.json'
```

```python
with open(file_name, 'r') as fout:
    results = json.load(fout)
df = pd.read_csv('active_hits_tweets.csv')
```

```python
dist = {}
for item in results:
    key = item['assignments_submitted_count']
    if key in dist:
        dist[key] = dist[key] + 1
    else:
        dist[key] = 1
        
dist
```

```python
arguments = []
for item in results:
    answers = item['answers']
    tweet = item['tweet']
    arg_dist = {}
    for answer in answers:
        key = answer['QuestionFormAnswers']['Answer']['FreeText'] 
        if key in arg_dist:
            arg_dist[key] = arg_dist[key] + 1
        else:
            arg_dist[key] = 1
    
    super_majority = max(list(arg_dist.values())) > 3
    arguments.append((tweet, arg_dist, super_majority))
        
```

```python
super_majors = 0

for (_, _, sup) in arguments:
    super_majors = super_majors + sup
super_majors
```

```python
for (tweet, arg_dist, sup) in arguments:
    row = df[df.tweet == tweet]
    df.loc[df.tweet == tweet, 'argumentative'] = (arg_dist['argumentative'] if 'argumentative' in arg_dist else 0)/sum(arg_dist.values())
    
```
